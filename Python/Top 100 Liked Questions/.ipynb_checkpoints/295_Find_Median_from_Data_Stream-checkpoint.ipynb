# Question

**reference:**<br>
[295. Find Median from Data Stream](https://leetcode.com/problems/find-median-from-data-stream/description/?envType=study-plan-v2&envId=top-100-liked)

The **median** is the middle value in an ordered integer list. If the size of the list is even, there is no middle value, and the median is the mean of the two middle values.

- For example, for `arr = [2,3,4]`, the median is `3`.
- For example, for `arr = [2,3]`, the median is `(2 + 3) / 2 = 2.5`.

Implement the MedianFinder class:

- `MedianFinder()` initializes the `MedianFinder` object.
- `void addNum(int num)` adds the integer `num` from the data stream to the data structure.
- `double findMedian()` returns the median of all elements so far. Answers within `10-5` of the actual answer will be accepted.

**Example 1:**

```python
Input
["MedianFinder", "addNum", "addNum", "findMedian", "addNum", "findMedian"]
[[], [1], [2], [], [3], []]
Output
[null, null, null, 1.5, null, 2.0]

Explanation
MedianFinder medianFinder = new MedianFinder();
medianFinder.addNum(1);    // arr = [1]
medianFinder.addNum(2);    // arr = [1, 2]
medianFinder.findMedian(); // return 1.5 (i.e., (1 + 2) / 2)
medianFinder.addNum(3);    // arr[1, 2, 3]
medianFinder.findMedian(); // return 2.0
```

**Constraints:**

- `-10^5 <= num <= 10^5`
- There will be at least one element in the data structure before calling `findMedian`.
- At most `5 * 10^4` calls will be made to `addNum` and `findMedian`.

## Min / Max Heap

Adding elements to `Heaps` takes logarithmic order of time, and gives direct access to the maximal / minimal elements in a group.

### Concept

A `Heap` is a special `Tree-based data structure` in which the tree is a complete `binary tree`. It is a natural fit for a priority queue. It enforces a partial order over elements, leading to fast enqueu, dequeue, and build times.

**Operations of Heap Data Structure**
- `Heapify`<br>
a process of creating a heap from an array impplementation of a tree;<br>
for every node **X**, the key of **X** is less than or equal to the key of **X**'s parent (except the root which has no parent), which means partial ordering but no ordering among siblings of a parent;
- `Insertion`<br>
a process to insert an element in existing heap time complexity **O(log N)**;<br>
do not need rotations! rotations are a consequence of total ordering;
- `Deletion`<br>
deleting the top element of the heap or the highest priority element, and then organizing the heap and returning the element with time complexity **O(log N)**;<br>
do not need rotations! rotations are a consequence of total ordering;
- `Peek`<br>
to check or find the first (or can say the top) element of the heap

**Types of Heap Data Structure**

Generally, heaps can be of two types，as shown below:
- 1. `Max Heap`<br>
In a `Max Heap` the key present at the root node must be greatest among the keys present at all of its children. The same property must be recursively true for all sub-trees in that `Binary Tree`.
- 2. `Min Heap`<br>
In a `Min Heap` the key present at the root node must be minimum among the keys present at all of its children. The same property must be recursively true for all sub-trees in that `Binary Tree`.

#### Difference between Min Heap and Max Heap

| Nr. | Min Heap | Max Heap |
| --- |:--- |:--- |
|1.   | The key present at the root node must be less than or equal to among the keys present at all of its children. | The key present at the root node must be larger than or equal to among the keys present at all of its children.|
|2.   |The minimum key element present at the root. | The maximum key element present at the root. |
|3.   |ascending priority | descending priority |
|4.   |The smallest element is the first to be popped from the heap. |The largest element is the first to be popped from the heap. | 

### Application

- `Heap Sort`<br>
`Heap Sort` is one of the best sorting algorithms that use `Binary Heap` to sort an array in **O(N*log N)** time;
- `Priority Queue`<br>
A priority queue can be implemented by using a heap because it supports **insert()**, **delete()**, **extractMax()**, **decreaseKey()** operations in **O(log N)** time;
- `Graph Algorithms`<br>
The heaps are especially used in Graph Algorithms like `Dijkstra's Shortest Path` and `Prim's Minimum Spanning Tree`.

### Concept

The word `binary` is generally associated with the number `2`. In this context, it refers to dividing a collection of elements into two halves and throwing away one of them at each step of the algorithm. This can dramatically reduce the number of comparisons required to find an element. But there’s a catch—elements in the collection must be sorted first.

The idea behind it resembles the steps for finding a page in a book. At first, you typically open the book to a completely random page or at least one that’s close to where you think your desired page might be.

Occasionally, you’ll be fortunate enough to find that page on the first try. However, if the page number is too low, then you know the page must be to the right. If you overshoot on the next try, and the current page number is higher than the page you’re looking for, then you know for sure that it must be somewhere in between.

You repeat the process, but rather than choosing a page at random, you check the page located right in the middle of that new range. This minimizes the number of tries. A similar approach can be used in the number guessing game. If you haven’t heard of that game, then you can look it up on the Internet to get a plethora of examples implemented in `Python`.

The page numbers that restrict the range of pages to search through are known as the lower bound and the upper bound. In binary search, you commonly start with the first page as the lower bound and the last page as the upper bound. You must update both bounds as you go. For example, if the page you turn to is lower than the one you’re looking for, then that’s your new lower bound.

Let’s say you were looking for a `strawberry` in a collection of fruits sorted in ascending order by size:



On the first attempt, the element in the middle happens to be a `lemon`. Since it’s bigger than a `strawberry`, you can discard all elements to the right, including the `lemon`. You’ll move the upper bound to a new position and update the middle index:



Now, you’re left with only half of the fruits you began with. The current middle element is indeed the `strawberry` you were looking for, which concludes the search. If it wasn’t, then you’d just update the bounds accordingly and continue until they pass each other. For example, looking for a missing `plum`, which would go between the `strawberry` and a `kiwi`, will end with the following result:



Notice there weren’t that many comparisons that had to be made in order to find the desired element. That’s the magic of `binary search`. Even if you’re dealing with a million elements, you’d only require at most a handful of checks. This number won’t exceed the logarithm base two of the total number of elements due to halving. In other words, the number of remaining elements is reduced by half at each step.

## Min / Max Heap

<img src="https://assets.leetcode.com/users/images/0eb8feba-cbfa-4f73-8d26-9aad226bdbc5_1626016093.9717174.png"  alt="ImageFile" style="width: 600px;" align="middle"/>

- [x] When dividing numbers into 2 balanced halves,<br>
- one max heap to keep the smaller half, the top of the max heap is the highest number among small numbers;<br>the other one min heap to keep the larger half，the top of the min heap is the lowest number among high numbers;
- we need to balance the size between max heap and min heap while processing. Hence, after adding `k` elements,<br>if k = 2 * i, `maxHeap.size` = `minHeap.size` = i;<br>if k = 2 * i + 1, `maxHeap.size` = `minHeap.size` + 1;

---

- [x] When adding a new number,<br>
- add `num` to the max heap, because :<br>
a new num should start with max heap that contains smaller half;<br>
make sure `maxHeap.size` larger than or equal to `minHeap.size`;
- move highest num in max heap to min heap, because :<br>
max heap may hold a big num, which should belong to min heap;<br>
balance `maxHeap.size` and `minHeap.size`;<br>
- move lowest num in min heap to max heap, because :<br>
min heap may hold more nums than max heap does;<br>
make sure `maxHeap.size` larger than or equal to `minHeap.size`; 

---

- [x] When finding the median,<br>
if `maxHeap.size` > `minHeap.size`，return the top of the `maxHeap`, which is the highest number among low numbers in the max heap;<br>
if `maxHeap.size` == `minHeap.size`，return the mean between `maxHeap.top()` and `minHeap.top()`);

**Complexity:**<br>

get min/max element: `O(1)`<br>
add new element: `O(logn)`<br>
remove min/max element: `O(logn)`<br>
add: `O(logn)`<br>
compute median: `O(1)`<br>

Total = `log1` + `log2` + ... + `logn` 
      = `O(nlogn)`

### Example

**Example 1:**

````python
import heapq

class MedianFinder:
    def __init__(self):
        self.minHeap = []                                                   # initialize min heap
        self.maxHeap = []                                                   # initialize max heap
        print(f">>> initialize")
        print(f"min heap: {self.minHeap}\nmax heap: {self.maxHeap}")

    def addNum(self, num: int) -> None:
        print()
        print(f">>> add num {num} to max heap")
        heapq.heappush(self.maxHeap, -num)                                  # add new num to max heap
        print(f"max heap: {self.maxHeap}\nmin heap: {self.minHeap}")
        print('-'*20)
        
        print(f"move the highest number in max heap to min heap")
        heapq.heappush(self.minHeap, -heapq.heappop(self.maxHeap))          # move the highest number in max heap to min heap
        print(f"max heap: {self.maxHeap}\nmin heap: {self.minHeap}")
        if len(self.minHeap) > len(self.maxHeap):                           # if minHeap.size > maxHeap.size,
            print(f"minheap.size: {len(self.minHeap)}\nmaxheap.size: {len(self.maxHeap)}")
            print(f"move the lowest number in min heap to max heap")
            heapq.heappush(self.maxHeap, -heapq.heappop(self.minHeap))      # move the lowest number in min heap to max heap, make sure 0< maxHeap.size - minHeap.size < 1
            print(f"min heap: {self.minHeap}\nmax heap: {self.maxHeap}")
            print('-' * 50)

    def findMedian(self) -> float:
        print()
        print(f">>> find the median")
        print(f"minheap.size: {len(self.minHeap)}\nmaxheap.size: {len(self.maxHeap)}")
        print(f"max heap: {self.maxHeap}\nmin heap: {self.minHeap}")
        if len(self.maxHeap) > len(self.minHeap):                           # if maxHeap.size > minHeap.size, return the highest number in max heap
            return -self.maxHeap[0]
        return (-self.maxHeap[0] + self.minHeap[0]) / 2                     # if maxHeap.size = minHeap.size, return the mean between the highest number in max heap and lowest number in min heap

obj = MedianFinder()
for i in range(2, 11):
    obj.addNum(i)
param_2 = obj.findMedian()
param_2
````

**Example 2:**

````python
import heapq

class MedianFinder:

    def __init__(self):
        self.count = 0                                                        # initialize count
        self.minHeap = []                                                     # initialize min heap
        self.maxHeap = []                                                     # initialize max heap
        print(f">>> initialize")
        print(f"count: {self.count}\nmin heap: {self.minHeap}\nmax heap: {self.maxHeap}")

    def addNum(self, num: int) -> None:
        self.count += 1

        print()
        print(f">>> add num {num} to max heap")
        heapq.heappush(self.maxHeap, (-num, num))                             # add new num to max heap
        print(f"max heap: {self.maxHeap}\nmin heap: {self.minHeap}")
        print('-'*20)

        print(f"move the highest number in max heap to min heap")
        _, max_heap_top = heapq.heappop(self.maxHeap)
        heapq.heappush(self.minHeap, max_heap_top)                            # move the highest number in max heap to min heap
        print(f"max heap: {self.maxHeap}\nmin heap: {self.minHeap}")
        print('-' * 50)
        
        if self.count & 1:                                                    # if len(nums) % 2 == 1:
            print(f"minheap.size: {len(self.minHeap)}\nmaxheap.size: {len(self.maxHeap)}")
            print(f"{self.count} % 2 = 1")
            print(f"move the lowest number in min heap to max heap")
            min_heap_top = heapq.heappop(self.minHeap)
            heapq.heappush(self.maxHeap, (-min_heap_top, min_heap_top))       # move the lowest number in min heap to max heap, make sure 0< maxHeap.size - minHeap.size < 1
            print(f"max heap: {self.maxHeap}\nmin heap: {self.minHeap}")
            print('-' * 50)

    def findMedian(self) -> float:
        print()
        print(f">>> find the median")
        print(f"minheap.size: {len(self.minHeap)}\nmaxheap.size: {len(self.maxHeap)}")
        print(f"max heap: {self.maxHeap}\nmin heap: {self.minHeap}")
        if self.count & 1:                                                    # if count is odd, return the highest number in max heap
            return self.maxHeap[0][1]
        else:                                                                 # if count is oven, return the mean between the highest number in max heap and lowest number in min heap
            return (self.minHeap[0] + self.maxHeap[0][1]) / 2

obj = MedianFinder()
nums = list(range(2, 25, 2))
print(f"nums: {nums}")
for i in nums:
    obj.addNum(i)
param_2 = obj.findMedian()
param_2
````

**Example 3:**

````python
import heapq

class MedianFinder:

    def __init__(self):
        self.minHeap = []                                                     # initialize min heap
        self.maxHeap = []                                                     # initialize min heap
        print(f">>> initialize")
        print(f"min heap: {self.minHeap}\nmax heap: {self.maxHeap}")

    def addNum(self, num: int) -> None:
        print()
        if len(self.minHeap) != len(self.maxHeap):                            # if maxHeap.size != minHeap.size
            print(f"minheap.size: {len(self.minHeap)}\nmaxheap.size: {len(self.maxHeap)}")
            print(f">>> add num {num} to min heap")
            heapq.heappush(self.minHeap, num)                                 # add new num to min heap
            print(f"max heap: {self.maxHeap}\nmin heap: {self.minHeap}")
            print('-' * 20)

            print(f"move the lowest number in min heap to max heap")
            heapq.heappush(self.maxHeap, -heapq.heappop(self.minHeap))        # move the lowest number in min heap to max heap
            print(f"max heap: {self.maxHeap}\nmin heap: {self.minHeap}")
            print('-' * 50)
        else:                                                                 # if maxHeap.size == minHeap.size
            print(f"minheap.size: {len(self.minHeap)}\nmaxheap.size: {len(self.maxHeap)}")
            print(f">>> add num {num} to max heap")
            heapq.heappush(self.maxHeap, -num)                                # add new num to max heap
            print(f"max heap: {self.maxHeap}\nmin heap: {self.minHeap}")
            print('-' * 20)

            print(f"move the highest number in max heap to min heap")
            heapq.heappush(self.minHeap, -heapq.heappop(self.maxHeap))        # move the highest number in max heap to min heap
            print(f"max heap: {self.maxHeap}\nmin heap: {self.minHeap}")
            print('-' * 50)

    def findMedian(self) -> float:
        print()
        print(f">>> find the median")
        print(f"minheap.size: {len(self.minHeap)}\nmaxheap.size: {len(self.maxHeap)}")
        print(f"max heap: {self.maxHeap}\nmin heap: {self.minHeap}")
        return self.minHeap[0] if len(self.minHeap) != len(self.maxHeap)\
            else (self.minHeap[0] - self.maxHeap[0]) / 2

obj = MedianFinder()
nums = list(range(2, 25, 2))
print(f"nums: {nums}")
for i in nums:
    obj.addNum(i)
param_2 = obj.findMedian()
print(f"param_2: {param_2}")
````

## Binary Search

**Complexity:**<br>

use two pointers to keep track the median position in the tree.

- when n is even, two pointers point to nearby middle elements;
- when n is odd, two pointers point to the same middle element;

add: `O(logn)`
- Insert into tree: `O(logn)`；
- Move pointers: `O(logn)`；
Compute median: `O(1)`

Total = log1 + log2 + ... + logn
      = O(nlogn)

### Example

````python
class MedianFinder:

    def __init__(self):
        self.nums = []                                                  # initialize nums
        print(f">>> initialize")
        print(f"nums: {self.nums}")

    def addNum(self, num: int) -> None:
        print()
        print(f"nums: {self.nums}")
        print(f">>> add num {num}")

        if not self.nums:                                                       # if currently nums is empty, directly append num to nums
            self.nums.append(num)
            print(f"append {num} to nums\nnums: {self.nums}\nreturn")
            return

        l = 0
        r = len(self.nums) - 1
        print(f"define 2 pointers\n{'l':20}: {l}\n{'r':20}: {r}")              # define 2 pointers, l(eft) and r(ight)
        while l < r:
            m = (l+r) // 2                                                      # define m pointed to median
            print(f"{'m':20}: {m}")
            if self.nums[m] == num:
                print(f"{'nums[m]=num':20}: nums[{m}] = {num}")
                self.nums.insert(m, num)
                print(f"{'nums[m]=num':20}: insert {num} to Nr.{m}\n{'nums':20}: {self.nums}")
                return
            elif self.nums[m] < num:                                            # if the median of nums smaller than num
                print(f"{'nums[m]<num':20}: nums[{m}] < {num}")
                print(f"{'l=m+1':20}: l={m}+1")
                l = m + 1                                                       # move l to the right of m
                print(f"{'l':20}: {l}\n{'r':20}: {r}")
            else:
                print(f"nums[{m}] > {num}")
                print(f"{'r=m-1':20}: r={m}-1")
                r = m - 1                                                       # move r to the left of m
                print(f"{'l':20}: {l}\n{'r':20}: {r}")
        print(f"compare nums[{l}] {self.nums[l]} and num {num}")
        if self.nums[l] < num:
            print(f"{'nums[l]<num':20}: nums[{l}] < {num}")
            self.nums.insert(l+1, num)                                          # add num to the right of l
            print(f"{'nums[l+1]=num':20}: insert {num} to Nr.{l+1}\n{'nums':20}：{self.nums}")
        else:
            print(f"{'nums[l]>num':20}: nums[{l}] > {num}")
            self.nums.insert(l, num)                                            # add num to the left of l
            print(f"{'nums[l]=num':20}: insert {num} to Nr.{l}\n{'nums':20}：{self.nums}")

    def findMedian(self) -> float:
        print()
        print(f">>> find the median")
        mid = (0 + len(self.nums)-1) // 2
        if (len(self.nums) - 1) & 1:
            return (self.nums[mid] + self.nums[mid+1]) / 2
        else:
            return self.nums[mid]

obj = MedianFinder()
nums = [5, 3, 15, 13, 11, 22, 20]
for i in nums:
    obj.addNum(i)
param_2 = obj.findMedian()
param_2
````
